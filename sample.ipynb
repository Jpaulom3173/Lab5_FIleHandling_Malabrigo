```python
# Step 1: Import additional libraries
from datetime import datetime
import shutil
# Step 2: Define working directory
backup_dir = Path.home() / "Documents" / "PythonFiles"
backup_dir.mkdir(exist_ok=True)
# Step 3: Define a function to write a file with backup
def write_with_backup(filename: str, content: str, mode="w"):
    file_path = backup_dir / filename
    
    if file_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_name = f"{file_path.stem}_Malabrigo_backup_{timestamp}{file_path.suffix}"
        backup_path = file_path.with_name(backup_name)
        shutil.copy2(file_path, backup_path)
        print(f"Backup saved: {backup_path.name}")
    
    with open(file_path, mode, encoding="utf-8") as file:
        file.write(content)
    print(f"File saved (Mode: {mode}): {file_path.name}")

def list_specific_backups(filename: str):
    file_stem = Path(filename).stem
    print(f"\nListing backups for {filename}:")
    for backup in backup_dir.glob(f"{file_stem}*backup*"):
        print("-", backup.name)
```

```python
from pathlib import Path
from datetime import datetime
import shutil

# --- SETUP ---
# Changing working directory to Surname_Activity_5
backup_dir = Path.home() / "Documents" / "Malabrigo_Activity_5"
backup_dir.mkdir(parents=True, exist_ok=True)

def file_manager():
    # Prompting for filename at the start; users can include extensions like .csv or .md
    print(f"Working Directory: {backup_dir.resolve()}")
    file_name = input("Enter filename with extension (e.g., data.csv or notes.md): ")
    file_path = backup_dir / file_name
    
    while True:
        print("\n--- MALABRIGO FILE MANAGER ---")
        print("1. Write to file (Overwrite)")
        print("2. Append to file")
        print("3. Read file")
        print("4. Backup file")
        print("5. Exit")
        
        
        choice = input("Choose an option (1-5): ")
        
        if choice == "1":
            content = input("Enter content to write:\n")
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(content)
            print(f"Success: Created/Overwrote {file_path.name}")
            
        elif choice == "2":
            more = input("Enter content to append:\n")
            with open(file_path, "a", encoding="utf-8") as f:
                f.write("\n" + more)
            print("Content added successfully.")
            
        elif choice == "3":
            if file_path.exists():
                with open(file_path, "r", encoding="utf-8") as f:
                    print(f"\n--- Content of {file_path.name} ---")
                    print(f.read())
            else:
                print("Error: File does not exist yet.")
                
        elif choice == "4":
            if file_path.exists():
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
                # Modified backup name to include Surname
                backup_file = file_path.with_name(
                    f"Malabrigo_{file_path.stem}_backup_{timestamp}{file_path.suffix}"
                )
                shutil.copy2(file_path, backup_file)
                print(f"Backup created: {backup_file.name}")
            else:
                print("Cannot backup. Original file not found.")
                
        elif choice == "5":
            print("Exiting. Happy Coding!")
            break
            
    
if __name__ == "__main__":
    file_manager()
```

```python
# Reverse File Content
from pathlib import Path
# Personal details
student_id = "2025-0355"
# Define file path
file_path = Path.home() / "Documents" /"Activity_5_Files" / f"lines_{student_id}.txt"
# Ensure file exists before processing
if not file_path.exists():
 sample_lines = ["Line 1", "Line 2", "Line 3"]
 file_path.write_text("\n".join(sample_lines))
 print(f"Sample file created for Student ID:{student_id}")
# Read, reverse, and overwrite the file
lines = file_path.read_text().splitlines()
lines.reverse()
with file_path.open("w") as f:
    f.write("\n".join(lines))
print(f"File lines reversed for Student ID:{student_id}")
```

```python
# Merge Two Files
from pathlib import Path
# Personal details
student_id = "2025-0355"
# Define file paths
documents_path = Path.home() / "Documents" /"Activity_5_Files"
f1 = documents_path / f"intro_{student_id}.txt"
f2 = documents_path / f"lines_{student_id}.txt"
merged = documents_path / f"merged_{student_id}.txt"
# Ensure both source files exist
if not f1.exists():
 f1.write_text(f"Welcome {student_id} to File Handling in Python!")
 print(f"Sample intro file created for Student ID:{student_id}")
if not f2.exists():
 f2.write_text("Line 1\nLine 2\nLine 3")
 print(f"Sample lines file created for Student ID:{student_id}")
# Merge contents of both files
with merged.open("w") as mf:
 mf.write(f1.read_text())
 mf.write("\n")
 mf.write(f2.read_text())
print(f"Files merged successfully for Student ID:{student_id}")
```
